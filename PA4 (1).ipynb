### NUMBER 2

#### Create a visualization that shows how the different features contributes to average grade.  Does chosen track in college, gender, or hometown contributes to a higher average score?


```python
#importing the matplotlib
import matplotlib.pyplot as plt
```

```python


# assigning the size of the plot
plt.figure(figsize=(10, 6))

# Create a box plot
df.boxplot(column='Average', by='Track', grid=True)

plt.xlabel('Track')
plt.ylabel('Average Score')
plt.title('Effect of College Track on Average Score')
plt.show()
```

```python
# Create a box plot for Gender vs Average
plt.figure(figsize=(10, 6))
track = plt.bar(df['Gender'], df['Average'])

plt.xlabel('Gender')
plt.ylabel('Average Score')
plt.title(' Gender on Average Score')
plt.show()
```

```python
# Create a box plot for Track vs Average
plt.figure(figsize=(10, 6))
track = plt.bar(df['Hometown'], df['Average'])

plt.xlabel('Hometown')
plt.ylabel('Average Score')
plt.title('Hometown on Average Score')
plt.show()
```
